```python
#give distribution of number of entries in each set
for key in sets_data.keys():
    print(key, len(sets_data[key]))
    print(np.bincount([len(x) for x in sets_data[key]]))

#plot distribution of number of entries in each set
for key in sets_data.keys():
    print(key, len(sets_data[key]))
    plt.hist([len(x) for x in sets_data[key]], bins=100)
    plt.show()
```
